```python
import requests
import json
s = requests.session()
x = s.get('https://v2.zekr.online/api/v1/quran/latest?lang=en')
s.cookies.clear()
json.loads(x.text)
```

```python
import time
def read_data_json(file_name):
    try:
        with open(file_name, 'r') as f:
            data= json.load(f)
    except:
        data = []
    return data
def save_data_json(file_name, new_data):
    try:
        with open(file_name, 'r') as f:
            data= json.load(f)
    except:
        data = []
    data = data + [new_data]
    with open(file_name, 'w+') as f:
        json.dump(data, f, indent=2)

def get_duration_of_zekr(base_url, file_name):
    data = read_data_json(file_name)
    if data == []:
        next_url = base_url
    else:
        next_url = data[-1]['next_page_url'] + '&lang=en&after_date=2000-01-01'
            
    total_duration =0
    while next_url != None:
        print('Next Url: ',next_url)
        s = requests.session()
        req = s.get(next_url, timeout=20)
        s.cookies.clear()
        
        req_dict = json.loads(req.text)
        current_page = req_dict['current_page']
        next_url = req_dict['next_page_url'] + '&lang=en&after_date=2000-01-01'  
        print('current_page= ',current_page)
        
        for item in req_dict['data']:
            total_duration += item['duration']
            
        save_data_json(file_name,req_dict)
        time.sleep(30)
            
    return total_duration
            
get_duration_of_zekr('https://v2.zekr.online/api/v1/quran/latest?page=1&lang=en&after_date=2000-01-01', 'zeker_metadata.json')
```
